Using SAS to access ADLS Gen 2 from Azure Databricks

```
# Making Spark configuration to access ADLS Gen 2 (aka Blob Storage) with SAS.
container_name = "PLEAES_ENTER_YOUR_OWNED_CONTAINER_NAME"
storage_account_name = "PLEAES_ENTER_YOUR_OWNED_STORAGE_ACCOUNT_NAME"
storage_account_to_connect = "fs.azure.sas."+container_name+"."+storage_account_name+".blob.core.windows.net"
sas_key = "PLEAES_ENTER_YOUR_OWNED_SAS_KEY"

spark.conf.set(storage_account_to_connect,sas_key)
```

```
# Using Databricks dbutils command to list folder in ADLS Gen 2 container or folder.
folder_name = ""
to_fs_ls = "wasbs://"+container_name+"@"+storage_account_name+".blob.core.windows.net/"+folder_name

dbutils.fs.ls(to_fs_ls)
```

```
# Spark read csv file in ADLS Gen 2 as dataframe.
df = spark.read.format("csv").load("wasbs://sample02@adls01.blob.core.windows.net/test1/gapminderDataFiveYear.csv")
df.printSchema()
```

```
# Display dataframe.
display(df)
```

```
# Spark read csv file in ADLS Gen 2 as dataframe, and mark first row as header.
df2 = spark.read.option("header",True).csv("wasbs://sample02@adls01.blob.core.windows.net/test1/gapminderDataFiveYear.csv")
df2.printSchema()
```

```
# Display dataframe.
display(df2)
```
